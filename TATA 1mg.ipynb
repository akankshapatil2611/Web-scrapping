```python
from bs4 import BeautifulSoup
import requests
from csv import writer
```

```python
ch_lst = []
alphabet = 'a'
for i in range(0,26):
    ch_lst.append(alphabet)
    alphabet = chr(ord(alphabet) + 1) 
print(ch_lst)
```

```python
r = range(2,22)
l1 = list()
for x in r :
    l1.append(x)
```

```python
def func(input_url):
    url = input_url
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    lists = soup.find_all('span', class_='Card__productCard__2MScM')
    with open('TATA1mg.csv','a', encoding='utf8', newline='') as f:
        thewriter = writer(f)
        header = ['MedicineName', 'Price','Salt Composition']
        thewriter.writerow(header)
        for list in lists:
            name = list.find('p', class_='Card__productName__nowyh bodyMedium').text
            composition = list.find_all('p', class_="Card__productDescription__-TaIB")[-1].text
            price = list.find('div', class_='Price__price__22Jxo').text
            info = [name, price, composition]
            thewriter.writerow(info)
```

```python
for ch in ch_lst:
    func("https://www.1mg.com/drugs-all-medicines?label={}".format(ch))
    print("Character {} and Page 1 written successfully".format(ch))
    for i in l1:
        ip = 'https://www.1mg.com/drugs-all-medicines?page={}&label={}'.format(i,ch)
        func(ip)
        print("Character {} and Page {} written successfully".format(ch,i))
```
